```python
#WC2003

from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests


url = "https://en.wikipedia.org/wiki/2003_Cricket_World_Cup"
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

team_r_03 =soup.find_all('td',style="width:47%; text-align:right; border-style:none")
team_l_03 =soup.find_all('td',style="width:47%; text-align:left; border-style:none")

right_teams_03 = []
for r in team_r_03:
    right_team = r.get_text(strip=True)
    right_teams_03.append(right_team)

left_teams_03 = []
for l in team_l_03:
    left_team = l.get_text(strip=True)
    left_teams_03.append(left_team)
```

```python
#WC2019

from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://en.wikipedia.org/wiki/2019_Cricket_World_Cup"
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

team_r_19 =soup.find_all('td',style="width:47%; text-align:right; border-style:none")
team_l_19 =soup.find_all('td',style="width:47%; text-align:left; border-style:none")

right_teams_19 = []
for r in team_r_19[10:]:
    right_team = r.get_text(strip=True)
    right_teams_19.append(right_team)

left_teams_19 = []
for l in team_l_19[10:]:
    left_team = l.get_text(strip=True)
    left_teams_19.append(left_team)
```

```python
team1= right_teams_03 + right_teams_07 + right_teams_11 + right_teams_15 + right_teams_19
```

```python
team2= left_teams_03 + left_teams_07 + left_teams_11 + left_teams_15 + left_teams_19
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame({"Team1":team1})
df2=pd.DataFrame({"Team2":team2})
```

```python
df1
```

```python
df1[['Team1','over1']] = df1['Team1'].str.split('(', 1, expand=True)
df1[['Over1','over1']] = df1['over1'].str.split(' ', 1, expand=True)
df1=df1.drop('over1', axis=1)
df1[['Team1','wicket1']] = df1['Team1'].str.split('/', 1, expand=True)
df1['Runs1'] = df1['Team1'].str.extract('(\d+)', expand=False)
df1['Team1'] = df1['Team1'].str.extract('([A-Za-z\s]+)', expand=False)
```

```python
df1
```

```python
df2
```

```python
df2[['Team2','over2']] = df2['Team2'].str.split('(', 1, expand=True)
df2[['Over2','over2']] = df2['over2'].str.split(' ', 1, expand=True)
df2=df2.drop('over2', axis=1)
df2[['Team2','wicket2']] = df2['Team2'].str.split('/', 1, expand=True)
df2['Runs2'] = df2['Team2'].str.extract('(\d+)', expand=False)
df2['Team2'] = df2['Team2'].str.extract('([A-Za-z\s]+)', expand=False)
```

```python
df2
```

```python
df=pd.concat([df1,df2],axis=1)
```

```python
df
```

```python
df=df[['Team1','Runs1','wicket1','Over1','Team2','Runs2','wicket2','Over2']]
```
